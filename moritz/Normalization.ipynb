```python
# fetch [0,1]

def normalize(frame, name):
        min_v = abs(min(frame[name].tolist()))
        max_v = abs(max(frame[name].tolist()))
        
        final = 0
        
        if max_v > min_v:
            final = max_v
        else:
            final = min_v
        
        # already in (0,1) -> skip
        if final <= 1.0:
            return frame
        
        for x in range(len(frame[name])):
            frame[name][x] = frame[name][x] / final
        return frame
```

```python
key_list = df.keys().tolist()
key_list.remove("ist_feiertag")
key_list.remove("ist_wohngebiet")
key_list.remove("ist_industriegebiet")
key_list.remove("datum")
key_list.remove("container_id")
key_list.remove("ist_ferien")
key_list
```

```python
# normalize values

for x in range(len(key_list)):
    df = normalize(df, key_list[x])
    
df.head()
```
